```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
frame = []
frame = pd.DataFrame(frame, columns=["Timestamp","X","Y", "Z", "File"]) 
rootdir = "/Users/Nick/Documents/University/Deakin/2023/SIT782 - Team Project (B) - Execution and Delivery/MobiFall_Dataset_v2.0/"
for subdirs, dirs, files in os.walk(rootdir):
    for file in files:
        path = os.path.join(subdirs, file)
        df = pd.read_csv(path, sep=",", names=["Timestamp","X","Y", "Z"])
        df['File'] = file
        if file == ".DS_Store":
            break
        else:
            for i in range(len(df['Timestamp'])):
                if df['Timestamp'][i] == "@DATA":
                    df = df[i+1:]
                    frame = frame.append(df)
                    break

frame['Timestamp'] = frame['Timestamp'].astype(int)
frame['X'] = frame['X'].astype(float)
frame['Y'] = frame['Y'].astype(float)
frame['Z'] = frame['Z'].astype(float)
frame['File'] = frame['File'].astype(str)

frame[['Movement', 'Measurement', 'Subject', 'Trial']] = frame['File'].str.split('_', expand=True)
frame['Trial'] = frame['Trial'].str.strip('.txt') 
frame = frame.drop(columns=['File'])
frame.to_csv('/Users/Nick/Downloads/raw_fall.csv', index=False)
```

```python
class1 = frame.copy(deep=True)
class1['Timestamp'] = round(class1['Timestamp']/100000000)
class1 = class1.sort_values(by=['Subject','Movement','Trial','Measurement','Timestamp'])
class1 = class1.drop_duplicates(subset=['Timestamp','Measurement'], keep='first')
class1 = class1.set_index(['Subject','Movement','Trial','Timestamp','Measurement'])
class1 = class1.unstack(level=-1).dropna()
class1['Fall'] = 0
class1 = class1.reset_index('Movement', drop=False)
class1['Fall'] = np.where(class1['Movement'] == "FOL",1,
                 np.where(class1['Movement'] == "FKL",1,
                 np.where(class1['Movement'] == "SDL",1,
                 np.where(class1['Movement'] == "BSC",1,0))))
class1 = class1.set_index('Movement', append=True)
```
